```python
chapiters = [f"{i:02}" if i < 10 else str(i) for i in range(1, 100)]
print(chapiters)
```

```python
import requests
from requests.adapters import HTTPAdapter
from requests.adapters import Retry
from bs4 import BeautifulSoup

def generate_urls(base_url, prefixes):
    urls = [f"{base_url}{prefix}" for prefix in prefixes]
    return urls

def scrape_href_tags_with_session(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }
    print(f"Fetching URL: {url}")
    session = requests.Session()
    retry = Retry(connect=3, backoff_factor=0.5)
    adapter = HTTPAdapter(max_retries=retry)
    session.mount('http://', adapter)
    session.mount('https://', adapter)
    response = session.get(url, headers=headers)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find all 'a' tags (anchor tags) and get their 'href' attributes
        href_tags = [a.get('href') for a in soup.find_all('a') if a.get('href')]

        return href_tags
    else:
        print("Failed to retrieve the webpage.")

# Generate URLs for different prefixes
base_url = 'https://www.douane.gov.ma/adil/ta_chap.asp?recherche1='
prefixes = chapiters # Add more prefixes as needed
urls = generate_urls(base_url, prefixes)

# Scrape href tags for each URL
href_tags_dict = {}
for i, url in enumerate(urls, start=1):
    href_tags = scrape_href_tags_with_session(url)
    href_tags_dict[f"href_tags{i}"] = href_tags

# Access the href_tags lists later using href_tags1, href_tags2, etc.
# print(href_tags_dict['href_tags1'])  # Access the href_tags from the first URL
# print(href_tags_dict['href_tags2'])  # Access the href_tags from the second URL
```

```python
def extract_position(href_tags_list):
    positions = []
    for href_tag in href_tags_list:
        position = href_tag.split('position=')[-1]
        positions.append(position)
    return positions

positions_dict = {}
for key, href_tags_list in href_tags_dict.items():
    positions_dict[key] = extract_position(href_tags_list)

# # Print the positions and their lengths for each href_tags list
# for key, positions in positions_dict.items():
#     print(f"Positions for {key}: {positions}")
#     print(f"Number of positions: {len(positions)}")
```

```python
def filter_positions(positions):
    filtered_positions = [pos for pos in positions if len(pos) == 10]
    return filtered_positions

filtered_positions_dict = {}
for key, positions_list in positions_dict.items():
    filtered_positions_dict[key] = filter_positions(positions_list)

# # Print the filtered positions and their lengths for each positions list
# for key, filtered_positions in filtered_positions_dict.items():
#     print(f"Filtered Positions for {key}: {filtered_positions}")
#     print(f"Number of filtered positions: {len(filtered_positions)}")
```

```python
def remove_duplicates(input_list):
    unique_list = list(set(input_list))
    return unique_list

# Create a dictionary to store unique lists for each filtered positions list
unique_lists_dict = {}

# Loop through filtered_positions_dict and apply the remove_duplicates function to each list
for i, (key, filtered_positions_list) in enumerate(filtered_positions_dict.items(), start=1):
    unique_lists_dict[f'unique_list{i}'] = remove_duplicates(filtered_positions_list)

    # If you want to explicitly create variables, uncomment the line below
    # locals()[f'unique_list{i}'] = remove_duplicates(filtered_positions_list)

    # Print the unique lists and their lengths for each filtered positions list
    print(f"Unique List for {key}: {unique_lists_dict[f'unique_list{i}']}")
    print(f"Number of unique elements: {len(unique_lists_dict[f'unique_list{i}'])}")
```

# # Designation Scraping (1, 2,... 99)

```python
import json
import requests
from bs4 import BeautifulSoup
from requests.adapters import Retry
from requests import Session
from requests.adapters import HTTPAdapter

def fetch_hierarchy(codes):
    parent_child_pairs = {}

    for code in codes:
        url = f"https://www.douane.gov.ma/adil/info_1.asp?pos={code}"
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
        }
        print(f"Fetching URL: {url}")
        session = requests.Session()
        retry = Retry(connect=3, backoff_factor=0.5)
        adapter = HTTPAdapter(max_retries=retry)
        session.mount('http://', adapter)
        session.mount('https://', adapter)
        response = session.get(url, headers=headers)
        if response.status_code != 200:
            print(f"Failed to fetch URL: {url}")
            break
        soup = BeautifulSoup(response.text, 'html.parser')
        tr_elements = soup.find_all('tr')[9:14]

        parent_child_pairs[code] = []
        parent = None

        # Iterate over the selected tr_elements
        for element in tr_elements:
            # Get the 5th td element
            td_elements = element.find_all('td')
            if len(td_elements) >= 5:
                child = td_elements[4].get_text(strip=True)
                # Remove "-" and whitespace
                child = child.replace("-", "").strip()
                parent_child_pairs[code].append(child)

    return parent_child_pairs

def save_hierarchy_to_json(hierarchy, filename):
    with open(filename, 'w') as json_file:
        json.dump(hierarchy, json_file, indent=4)

def fetch_and_save_hierarchies(unique_lists_dict, base_filename):
    hierarchies = {}
    for i, codes in unique_lists_dict.items():
        hierarchy = fetch_hierarchy(codes)
        hierarchies[f'chapter{i}'] = hierarchy
        save_hierarchy_to_json(hierarchy, f"{base_filename}{i}.json")
    return hierarchies

# Example of usage:
# Assuming unique_lists_dict is defined and contains the unique lists for all chapters

# Fetch and save hierarchies for all chapters
all_hierarchies = fetch_and_save_hierarchies(unique_lists_dict, 'hierarchy')

# You can use all_hierarchies for further processing if needed
```

## Filtering (1st and last)

```python
import json

def retrieve_first_and_last(json_file_path, output_file_path):
    # Load the JSON data
    with open(json_file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)

    # Create a new dictionary to store modified data
    modified_data = {}

    # Iterate over each key-value pair
    for key, value in data.items():
        # Check if the list has at least one element
        if value:
            # Retrieve the first and last elements in the list
            first_and_last = [value[0], value[-1]]
        else:
            # If the list is empty, assign None for both first and last
            first_and_last = [None, None]
        # Assign the new value to the modified dictionary
        modified_data[key] = first_and_last

    # Save the updated JSON data
    with open(output_file_path, 'w', encoding='utf-8') as file:
        json.dump(modified_data, file, indent=4)

def process_all_files(base_filename, num_files):
    for i in range(1, num_files + 1):
        input_file = f'C:\\Users\\admin\\hierarchyunique_list\\{base_filename}{i}.json'
        output_file = f'FilteredChapter{i}.json'
        retrieve_first_and_last(input_file, output_file)

# Example of usage:
base_filename = 'hierarchyunique_list'
num_files = 99
process_all_files(base_filename, num_files)
```

## Cleaning 

```python
import json

def load_and_decode_json(filename):
    with open(filename, 'r', encoding='utf-8') as json_file:
        data = json.load(json_file)
    
    # Decode encoded strings
    decoded_data = {}
    for key, value in data.items():
        decoded_key = key.encode('latin1').decode('utf-8')
        decoded_values = [v.encode('latin1').decode('utf-8') for v in value]
        decoded_data[decoded_key] = decoded_values

    return decoded_data

def save_as_json(data, filename):
    with open(filename, 'w', encoding='utf-8') as json_file:
        json.dump(data, json_file, ensure_ascii=False, indent=4)

# Example usage:
input_filename = 'hierarchyFiltered29.json'
output_filename = 'newhierarchyCleaned29.json'

decoded_data = load_and_decode_json(input_filename)
save_as_json(decoded_data, output_filename)
```

```python
import json

def replace_unicode_escape(input_file_path, output_file_path):
    # Read input JSON file
    with open(input_file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)

    # Perform replacements on each record
    for key, value in data.items():
        if isinstance(value, list):
            new_value = []
            for item in value:
                # Perform replacements on each item in the list
                new_item = item.replace("", "").replace("", "").replace("", "").replace("", "i").replace("","").replace("", "").replace("", "")
                new_value.append(new_item)
            data[key] = new_value
    # Save the modified JSON to the output file
    with open(output_file_path, 'w', encoding='utf-8') as file:
        json.dump(data, file, indent=4, ensure_ascii=False)


input_file_path = 'hierarchyFiltered29.json'
output_file_path = 'hierarchyCleaned29.json'
replace_unicode_escape(input_file_path, output_file_path)


# input_file_path = 'hierarchyFiltered28.json'
# output_file_path = 'hierarchyCleaned28.json'
# replace_unicode_escape(input_file_path, output_file_path)

# input_file_path = 'hierarchyFiltered27.json'
# output_file_path = 'hierarchyCleaned27.json'
# replace_unicode_escape(input_file_path, output_file_path)
```
