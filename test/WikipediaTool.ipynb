```python
import re

def clean_nomenclature(nomenclature):
    cleaned_list = []
    for item in nomenclature:
        # Remove everything after "et"
        cleaned_item = item.split(" et ")[0].strip()
        # Remove content within parentheses
        cleaned_item = re.sub(r'\([^()]*\)', '', cleaned_item).strip()
        # Erase connecting words (leurs..etc)
        connecting_words = ["leurs", "et", "et ses", "ses", "de", "sels", "LEURS"]
        for word in connecting_words:
            cleaned_item = cleaned_item.replace(word, "")
        # Remove content after "Dérivés"
        cleaned_item = cleaned_item.split(" dérivés ")[0].strip()
        cleaned_list.append(cleaned_item)
    return cleaned_list

cleaned_nomenclature = clean_nomenclature(nomenclature)
for item in cleaned_nomenclature:
    print(item)
```

```python
import wikipedia

def find_related_pages(nomenclature_list):
    related_words = set()  # Use a set to store unique words
    processed_words = set()  # Use another set to store processed words
    for item in nomenclature_list:
        words = item.split()
        for word in words:
            # Check if the word has already been processed
            if word not in processed_words:
                processed_words.add(word)  # Add the word to the processed set
                # Check if the word has already been found to have a related Wikipedia page
                if word not in related_words:
                    try:
                        page = wikipedia.page(word)
                        related_words.add(word)  # Add the word to the related set
                        print(f"Found related page for '{word}': {page.url}")
                    except wikipedia.exceptions.PageError:
                        print(f"No related page found for '{word}'")
                    except wikipedia.exceptions.DisambiguationError as e:
                        print(f"Ambiguous term '{word}', please choose one: {e.options}")
    return list(related_words)  # Convert the set back to a list

related_pages = find_related_pages(cleaned_nomenclature)
print("Words with related Wikipedia pages:", related_pages)
```
